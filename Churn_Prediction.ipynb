**DATA PRE PROCESSING**

```
datasets["MonthlyCharges"] = pd.to_numeric(datasets["MonthlyCharges"], downcast="float")
```

```
datasets.head()
```

```
datasets=datasets.drop(['customerID'], axis=1)
```

**TRAINING AND TESTING SET**

```
Y=datasets.iloc[:,-1].values
```

```
X=datasets.iloc[:,[1,4,17]]
```

```
X
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.30)
```

**LOGISTIC REGRESSION**

```
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(fit_intercept = False, C = 1e12, solver 
= 'liblinear')

logreg.fit(X_train, Y_train)
```

```
y_hat_train = logreg.predict(X_train)
y_hat_test = logreg.predict(X_test)
```

**PERFORMANCE METRICS OF LOGISTIC REGRESSION**

```
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
result = confusion_matrix(Y_test, y_hat_test)
print("Confusion Matrix:")
print(result)
result1 = classification_report(Y_test, y_hat_test)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(Y_test,y_hat_test)
print("Accuracy:",result2)
```
